```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
rows=[]
for page_num in range(1,11):
    url=f'https://www.epicurious.com/search/cucumbers?{page_num}'
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    recipes=doc.find_all(class_='recipe-content-card')
    for recipe in recipes:
        row={}
        try:
            row['tag']=recipe.find(class_='tag').text.strip()
        except:
            pass
        try:
            row['title']=recipe.find('h4').text.strip()
        except:
            pass
        try:
            row['summary']=recipe.find(class_='dek').text.strip()
        except:
            pass
        try:
            row['rating']=recipe.find(class_='rating').find('span').text.strip()
        except:
            pass
        try:
            row['wma']=recipe.find(class_='make-again-percentage').text.strip()
        except:
            pass
        try:
            row['url']=recipe.find(class_='hed').find('a')['href']
        except:
            pass
        rows.append(row)
print(rows)
```

```python
df=pd.DataFrame(rows)
df
```

```python
df=df[df['tag'] == 'recipe']
df
```

```python
def scrape_page(row):
    url = f"https://www.epicurious.com" + row['slug']
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    page={}
    try:
        page['ingrdients']=doc.find(class_='ingredient-groups').text.strip()
    except:
        pass
    try:
        page['directions']=doc.find(class_='preparation-groups').text.strip()
    except:
        pass
    try:
        page['tags']=doc.find(class_='tags').text.strip()
    except:
        pass
    return pd.Series(page)
```

```python
scrape_page(df.loc[0])
```

```python
scraped_df = df.apply(scrape_page, axis=1)
scraped_df
```

```python
df_merged=df.merge(scraped_df, left_index=True, right_index=True)
```
