```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors"
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
books=doc.find_all('tr')
books
```

```python
rows=[]
for book in books:
    row={}
    row['rank']=book.find(class_='number').text.strip()
    row['title']=book.find('span').text.strip()
    row['author']=book.find_all('span')[2].text.strip()
    row['score']=book.find(class_='smallText uitext').find('a').text.strip()
    row['vote']=book.find(class_='smallText uitext').find_all('a')[1].text.strip()
    row['rating']=book.find(class_='minirating').text.strip()
    rows.append(row)
df=pd.DataFrame(rows)
df = df[['rank','title','author','score','vote','rating']]
df.head()

    
    
```

```python
df['score']=df.score.str.replace('score:',"")
df
```

```python
rating_cleaned=[]
for rate in df.rating:
    rate_cleaned=rate.split(' ')[0]
    rating_cleaned.append(rate_cleaned)
df['rating']=rating_cleaned
df
```

```python
title_cleaned=[]
for t in df.title:
    t_cleaned=t.split('(')[0]
    title_cleaned.append(t_cleaned)

df['title_cleaned']=title_cleaned
df
```

```python
s_n=[]
for t in df.title:
    if "(" in t:
        sn_cleaned=t.split('(')[1]
        s_n.append(sn_cleaned)
    else:
        s_n.append('/')
df['s_n']=s_n
df

```

```python
series_cleaned=[]
numbers_cleaned=[]
for s in df.s_n:
    if ')' in s:
        series=s.split('#')[0]
        numbers=s.split('#')[1]
        series_cleaned.append(series)
        numbers_cleaned.append(numbers)
    else:
        series_cleaned.append('/')
        numbers_cleaned.append('/')
df['series_cleaned']=series_cleaned
df['numbers_cleaned']=numbers_cleaned
df
```

```python
sc_cleaned=[]
for sc in df.series_cleaned:
    if ',' in sc:
        scc=sc.replace(',','')
        sc_cleaned.append(scc)
    else:
        scc=sc
        sc_cleaned.append(scc)
df['sc_cleaned']=sc_cleaned
df

    
```

```python
ncc_cleaned=[]
for nc in df.numbers_cleaned:
    if ')' in nc:
        ncc=nc.replace(')','')
        ncc_cleaned.append(ncc)
    else:
        ncc=nc
        ncc_cleaned.append(ncc)
df['ncc_cleaned']=ncc_cleaned
df

    
```
